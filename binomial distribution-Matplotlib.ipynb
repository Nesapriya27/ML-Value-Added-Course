```python
#Binomial distribution or Discrete distribution 
Ex:Toss of a coin,it has 3 arguments 
    1 No of trails(n)
    2 Probability of occurence of each trials(p=0.5 each)
    3 Size of the resultant array
```

```python
from numpy import random
a=random.binomial(n=10,p=0.5,size=10)
print(a)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
a=random.binomial(n=10,p=0.5,size=1000)
sb.distplot(a,kde=False)
plt.show()
```

```python
a=random.binomial(n=100,p=0.5,size=1000)
b=random.normal(loc=50,scale=5,size=1000)
sb.distplot(a,hist=False)
sb.distplot(b,hist=False)
plt.show()
```

```python
#Poisson distribution

It is also a discrete distribution
It has 2 attributes
known number of occurences/rate of occurence(lam)
Shape of the resultant array(size)
```

```python
a=random.poisson(lam=3,size=30)
print(a)
```

```python
a=random.poisson(lam=3,size=1000)
sb.distplot(a,kde=False)
plt.show()
```

```python
#Uniform distribution

The event has equal chances of occuring
It has 3 paramters
1.Lower bound(a=0)#default
2.Upper bound(a=1)#default 
3.Shape of the array(size)
```

```python
a=random.uniform(size=(3,3))
print(a)
```

```python
a=random.uniform(size=(3,3))
sb.distplot(a,hist=False)
plt.show()
```

```python
sb.distplot(a,hist=False)
plt.show()
```

```python
#Logistic distribution

It is used in Machine learning for logistic regression
It is used in neural networks
It is used to represent the growth
It has 3 attributes
1.loc-Mean(default 0)
2.Scale-SD(default 1) 
3.shape of the array(size)
```

```python
a=random.logistic(loc=1,scale=2,size=1000)
print(a)
```

```python
sb.distplot(a,hist=False)
plt.show()
```

```python
a=random.logistic(loc=1,scale=2,size=1000)
b=random.normal(loc=1,scale=5,size=1000)
sb.distplot(a,hist=False,label="Logistic",color="red")
sb.distplot(b,hist=False,label="Normal",color="green")

```

```python
#Multinomial distribution

It is also known as polynomial distribution
Ex:Rolling a dice
It has 3 parameters
1.n-No of outcomes 
2.pvals-List of probabilities
3.shape of the array(Size)
```

```python
a=random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6])
print(a)
```

```python
a=random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=10)
print(a)
```

```python
sb.distplot(a)
plt.show()
```

```python
#Exponential distribution

Time till next event
It has two parameters
1.Inverse of rate of occurence(default=1)
2.Size
```

```python
a=random.exponential(size=1000)
sb.distplot(a,hist=False)
plt.show()
```

```python
Chi square distribution

To verify hypothesis
It has 2 arguments 
1.df(Degree of freedom) 
2.shape of the array(size)
```

```python
a=random.chisquare(df=1,size=1000)
print(a)
```

```python
sb.distplot(a)
plt.show()
```

```python
#Rayleigh distribution

It is used for signal processing
It has 2 arguments
1.Scale(SD)#default=1
2.Size
```

```python
a=random.rayleigh(size=1000)
sb.distplot(a,hist=False)
plt.show()
```

```python
#Pareto distribution

It is known as 80-20 distribution
20% factors cause 80% outcomes
It has 2 arguments 
1.Shape of the argument 
2.Size
```

```python
a=random.pareto(a=2,size=1000)
sb.distplot(a,hist=False)
plt.show()
```

```python
#zipf distribution zips law

In a collection nth common term is 1/n times of the most common term
Ex:Fifth most common word in English occurs
nearly 1/5 times often as most common word
It has 2 arguments 
1.a(distribution parameter) 
2.size
```

```python
a=random.zipf(a=2,size=1000)
sb.distplot(a[a<10],kde=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
plt.scatter(x1,y1)
plt.title('India')
plt.show()
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
plt.scatter(x1,y1)
x2=np.array([1,2,3,4,5])
y2=np.array([10,7,16,19,2])
plt.scatter(x2,y2)
plt.title('Ind vs Aus')
plt.show()
plt.scatter(x1,y1,c='r')
plt.scatter(x2,y2,c='g')
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
c1=np.array(['red','green','blue','orange','magenta'])
plt.scatter(x1,y1,c=c1)
x2=np.array([1,2,3,4,5])
y2=np.array([10,7,16,19,2])
c2=np.array(['cyan','black','grey','brown','pink'])
plt.scatter(x2,y2,c=c2)
plt.title('Ind vs Aus')
plt.show()
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
c1=np.array(['red','green','blue','orange','magenta'])
plt.scatter(x1,y1,c=c1,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x1,y1,c=c1,cmap='Accent')

```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors)


plt.show()
```

```python
#increase the size
x = np.random.randint(10, size=(10))
y = np.random.randint(10, size=(10))
colors = np.random.randint(10, size=(10))
size=np.array([10,15,20,50,100,30,40,25,55,60])
plt.scatter(x, y, c=colors,s=size)
plt.show()
```

```python
#alpha-denotes the opacity
plt.scatter(x, y, c=colors,s=size,alpha=0.5)
```
