# Pandas


```python
Pandas is a Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.

The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

Data Science: is a branch of computer science where we study how to store, use and analyze data for deriving information from it.

Pandas allows us to analyze big data and make conclusions based on statistical theories.
Pandas can clean messy data sets, and make them readable and relevant.

Relevant data is very important in data science.
```

# panda series

```python
Its like a 1d array holding data of any type
Its like a column in a table
Here the indices are used as label.
Usually the index starts from 0.
Index is used to access specific value from the series
```

```python
ds=[90,95,89,65,88]
df=pd.Series(ds)
print(df)
```

```python
print(df[2])
print(df[0])
print(df['89'])# should be accessed only by index and not by value
```

```python
ds=[90,95,89,65,88]
df=pd.Series(ds,index=['T','E','M','S','SS'])
print(df)
```

```python
print(df[1])
```

```python
print(df['T'])
```

```python
print(df['E'])
```

```python
ds={'T':90,'E':95,'M':89,'S':65,'SS':88}
df=pd.Series(ds)
print(df)
```

# DataFrame

```python
It is a multidimensional table in which each column is a series.
```

```python
import pandas as pd
ds={'names':['Tushar','Saran','Madhan','Sabari'],'marks':[95,89,93,98]}
x=pd.DataFrame(ds)
print(x.loc[0])
print(x.loc[[1,2]])
```

```python
ds={'Tam':[90,86,98,100,87],'Eng':[95,89,93,98,79],'Mat':[89,95,79,90,100]}
df=pd.DataFrame(ds,index=['Test1','Test2','Test3','Test4','Test5'])
print(df)
```
